## Consigna:
En este primer trabajo comenzaremos por diseñar un generador de señales que utilizaremos en las primeras simulaciones que hagamos. La primer tarea consistirá en programar una función que genere señales senoidales y que permita parametrizar:

    la amplitud máxima de la senoidal (volts)
    su valor medio (volts)
    la frecuencia (Hz)
    la fase (radianes)
    la cantidad de muestras digitalizada por el ADC (# muestras)
    la frecuencia de muestreo del ADC.

es decir que la función que uds armen debería admitir se llamada de la siguiente manera

tt, xx = mi_funcion_sen( vmax = 1, dc = 0, ff = 1, ph=0, nn = N, fs = fs)

## Resolucion:

```python
import numpy as np
import matplotlib.pyplot as plt

def mi_funcion_sen(vmax=1, dc=0, ff=1, ph=0, nn=100, fs=1000):
    """
    Funcion:
        Realiza una senoidal con los parametros ingresados
        
    Parámetros:
        vmax: Amplitud máxima de la senoidal
        dc: Valor medio
        ff: Frecuencia de la señal
        ph: Fase de la señal (radianes)
        nn: Cantidad de muestras
        fs: Frecuencia de muestreo
    
    Retorna:
        tt: Vector de tiempos
        xx: Vector de valores de la señal
    """
    tt = np.arange(nn) / fs
    xx = vmax * np.sin(2 * np.pi * ff * tt + ph) + dc
    
    return tt, xx
```

## Ejemplo de uso:
A continuacion se mostrara la utilizacion de la funcion desarrollada.

```python
tt, xx = mi_funcion_sen(vmax=1, dc=0.5, ff=5, ph=np.pi/2, nn=1000, fs=5000)

# Graficar la señal
plt.plot(tt, xx)
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (V)")
plt.title("Señal Senoidal Generada")
plt.grid()
plt.show()
```

## Casos especiales:
Para el uso correcto de esta función se recomienda utilizar una frecuencia menor a la mitad de la frecuencia de muestreo, respetando así el teorema de Nyquist. De lo contrario, la señal original se verá afectada por aliasing, donde su espectro se superpone y deforma la reconstrucción de la señal. Además, se evaluarán algunos casos especiales de frecuencia para observar el efecto del espectro periódico propio del muestreo.

#### Frecuencia igual a N veces la frecuencia de muestreo:
Cuando la frecuencia de la señal es un múltiplo exacto de la frecuencia de muestreo, se observa una línea horizontal o una señal de frecuencia aparente igual a cero.
Esto sucede porque, debido a la periodicidad espectral introducida por el muestreo, todas las copias del espectro se repiten a intervalos de $f_{s}$​ (frecuencia de muestreo). Así, las componentes de frecuencia coinciden en cero al superponerse, provocando que la señal parezca constante.

```python
fs = 1000
tt, xx = mi_funcion_sen(vmax=1, dc=0.5, ff=fs*3, ph=np.pi/2, nn=1000, fs=fs)

# Graficar la señal
plt.plot(tt, xx)
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (V)")
plt.title("Señal Senoidal Generada")
plt.grid()
plt.show()
```

#### Frecuencia cercanas a $\frac{f_{s}}{2}$:
Cuando se eligen frecuencias próximas a la mitad de la frecuencia de muestreo, el espectro de la señal se pliega sobre sí mismo. Este fenómeno es similar a lo que se estudia en comunicaciones como el fenómeno de aliasing o plegamiento de espectro (similar a un gráfico de Banda Lateral Única o BLU).
En esta situación, el dispositivo de muestreo no puede diferenciar correctamente entre las componentes de alta y baja frecuencia, lo que produce un gráfico distorsionado donde las señales se solapan.

```python
fs = 1000
ff = fs/2-1
tt, xx = mi_funcion_sen(vmax=1, dc=0.5, ff=ff, ph=0, nn=1000, fs=fs)

# Graficar la señal
plt.plot(tt, xx)
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (V)")
plt.title("Señal Senoidal Generada")
plt.grid()
plt.show()
```
