<p align="center">
    \( X_k = \sum_{n=0}^{N-1} x_n \, e^{-j \frac{2\pi k n}{N}} \)

XX = mi_funcion_DFT( xx )

xx: señal a analizar, una matriz (Nx1) de números reales. 

XX: DFT de xx, una matriz (Nx1) de números complejos.

## Resolucion:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mi_funcion_DFT(xx):
    """
    Funcion:
        Calcula la DFT de una señal de entrada xx.

    Parámetros: 
        xx : Señal de entrada (N x 1), números reales.

    Retorna:
        XX : DFT de la señal (N x 1), números complejos.
    """
    N = len(xx)
    n = np.arange(N)
    k = n.reshape((N, 1))
    W = np.exp(-2j * np.pi * k * n / N)
    XX = np.dot(W, xx)
    return XX
```

```python
# Funcion realizada en Tp_Semana_1
def mi_funcion_sen(vmax=1, dc=0, ff=1, ph=0, nn=100, fs=1000):
    """
    Funcion:
        Realiza una senoidal con los parametros ingresados
        
    Parámetros:
        vmax: Amplitud máxima de la senoidal
        dc: Valor medio
        ff: Frecuencia de la señal
        ph: Fase de la señal (radianes)
        nn: Cantidad de muestras
        fs: Frecuencia de muestreo
    
    Retorna:
        tt: Vector de tiempos
        xx: Vector de valores de la señal
    """
    tt = np.arange(nn) / fs
    xx = vmax * np.sin(2 * np.pi * ff * tt + ph) + dc
    
    return tt, xx
```

## Ejemplo de uso:
#### Señal de 62.5Hz

```python
# Parámetros comunes
fs = 1000  # frecuencia de muestreo en Hz
N = 128    # cantidad de muestras

# Senoidal con frecuencia ajustada (62.5 Hz)
tt, xx_2 = mi_funcion_sen(ff=62.5, nn=N, fs=fs)

# Calcular DFT y FFT
DFT_xx_2 = mi_funcion_DFT(xx_2)
FFT_xx_2 = np.fft.fft(xx_2)

# Frecuencias, utilizo el mismo vector que para el tiempo pero lo ajusto para que tenga un correcto espaciamiento
frequencies = tt*fs**2/N

# Plot senoidal ajustada
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.stem(frequencies, np.abs(DFT_xx_2), basefmt=" ")
plt.title("DFT de senoidal ajustada (62.5 Hz)")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, fs/2)
plt.grid()

plt.subplot(1, 2, 2)
plt.stem(frequencies, np.abs(FFT_xx_2), basefmt=" ")
plt.title("FFT de senoidal ajustada (62.5 Hz)")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, fs/2)
plt.grid()
plt.tight_layout()
plt.show()
```

## Caso especial:
La DFT asume que la señal que analizás es periódica en el intervalo de longitud N, y sólo puede detectar frecuencias que coincidan exactamente con los multiplos de la resolucion espectral, por lo cual si se desea obtener un delta como en el caso anterior se debe cumplir lo siguiente:

$$f_{f}=\alpha .\Delta f=\frac{\alpha . f_{s}}{N} $$
$$\alpha \rightarrow \mathbb{N}$$

Lo mas normal cuando intentemos cuando realisemos este tipo de calculos es que la frecuencia muestreada no coincida con la resolucion espectral en esos casos el espectro quedara disperso como se muestra en el siguiente caso.

### Señal de 50 Hz

```python
# Senoidal de 50 Hz
tt, xx_1 = mi_funcion_sen(ff=50, nn=N, fs=fs)

# Calcular DFT y FFT
DFT_xx_1 = mi_funcion_DFT(xx_1)
FFT_xx_1 = np.fft.fft(xx_1)

# Frecuencias, utilizo el mismo vector que para el tiempo pero lo ajusto para que tenga un correcto espaciamiento
frequencies = tt*fs**2/N

# Plot senoidal original
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.stem(frequencies, np.abs(DFT_xx_1), basefmt=" ")
plt.title("DFT de senoidal (50 Hz)")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, fs/2)
plt.grid()

plt.subplot(1, 2, 2)
plt.stem(frequencies, np.abs(FFT_xx_1), basefmt=" ")
plt.title("FFT de senoidal (50 Hz)")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, fs/2)
plt.grid()
plt.tight_layout()
plt.show()
```

## Ruido:
A continuacion se mostrara como quedara la DFT de un ruido unirforma de variaza de 4

```python
# Ruido uniforme con varianza 4
# varianza de ruido uniforme entre [a, b] es (b-a)^2 / 12 => para var=4, b-a = sqrt(48) ≈ 6.93
varianza = 4
a = -np.sqrt(12 * varianza) / 2
b = np.sqrt(12 * varianza) / 2

ruido = np.random.uniform(a, b, N)

# Calcular DFT y FFT
DFT_ruido = mi_funcion_DFT(ruido)
FFT_ruido = np.fft.fft(ruido)

# Plot ruido en tiempo
plt.figure(figsize=(12, 4))
plt.plot(ruido)
plt.title("Ruido uniforme (var=4)")
plt.xlabel("Muestra")
plt.ylabel("Valor")
plt.grid()

# Plot ruido
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.stem(frequencies, np.abs(DFT_ruido), basefmt=" ")
plt.title("DFT de ruido uniforme (var=4)")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, fs/2)
plt.grid()

plt.subplot(1, 2, 2)
plt.stem(frequencies, np.abs(FFT_ruido), basefmt=" ")
plt.title("FFT de ruido uniforme (var=4)")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, fs/2)
plt.grid()
plt.tight_layout()
plt.show()
```

#### Ruido + Señal:
Aprobechando que la longitud de la señal es igual a la longitud del ruido la sumaremos en tiempo para ver su comportamiento en en frecuencia, para eso utilizaremos la primera señal mostrada (62.5 Hz) para poder identificarla mas facilmente.

```python
rxx_2 = ruido + xx_2

# Calcular DFT y FFT
DFT_rxx_2 = mi_funcion_DFT(rxx_2)
FFT_rxx_2 = np.fft.fft(rxx_2)

# Plot ruido en tiempo
plt.figure(figsize=(12, 4))
plt.plot(rxx_2)
plt.title("Señal con ruido uniforme (var=4)")
plt.xlabel("Muestra")
plt.ylabel("Valor")
plt.grid()

# Plot ruido
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.stem(frequencies, np.abs(DFT_rxx_2), basefmt=" ")
plt.title("DFT de señal ruidosa")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, fs/2)
plt.grid()

plt.subplot(1, 2, 2)
plt.stem(frequencies, np.abs(FFT_rxx_2), basefmt=" ")
plt.title("FFT de señal ruidosa")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, fs/2)
plt.grid()
plt.tight_layout()
plt.show()
```

Como se ve en el grafico de arriba el ruido es muy alto por lo cual no se puede apreciar correctamente la señal, debido a esto se tomo la decicion de aumentar 10dB el señal de entrada para poder apreciar mejor los graficos.

```python
rxx_2 = ruido + xx_2*10

# Calcular DFT y FFT
DFT_rxx_2 = mi_funcion_DFT(rxx_2)
FFT_rxx_2 = np.fft.fft(rxx_2)

# Plot ruido en tiempo
plt.figure(figsize=(12, 4))
plt.plot(rxx_2)
plt.title("Señal con ruido uniforme (var=4)")
plt.xlabel("Muestra")
plt.ylabel("Valor")
plt.grid()

# Plot ruido
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.stem(frequencies, np.abs(DFT_rxx_2), basefmt=" ")
plt.title("DFT de señal ruidosa")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, fs/2)
plt.grid()

plt.subplot(1, 2, 2)
plt.stem(frequencies, np.abs(FFT_rxx_2), basefmt=" ")
plt.title("FFT de señal ruidosa")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Magnitud")
plt.xlim(0, fs/2)
plt.grid()
plt.tight_layout()
plt.show()
```
