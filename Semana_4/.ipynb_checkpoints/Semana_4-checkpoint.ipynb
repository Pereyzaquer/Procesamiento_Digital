### TS4: Simulaci칩n de un ADC (cuantizaci칩n y alias) parte 2/2

En esta tarea semanal retomamos la consigna de la tarea anterior, donde simulamos el bloque de cuantizaci칩n de un ADC de B bits en un rango de  췀VF Volts. Ahora vamos a completar la simulaci칩n del ADC incluyendo la capacidad de muestrear a fs Hertz.

Para ello se simular치 el comportamiento del dispositivo al digitalizar una senoidal contaminada con un nivel predeterminado de ruido. Comenzaremos describiendo los par치metros a ajustar de la senoidal:

- Frecuencia $f_{o}$ arbitraria, por ejemplo $f_{0}=\frac{f_{S}}{N}=\Delta f$
- Energ칤a normalizada, es decir energ칤a (o varianza) unitaria

Con respecto a los par치metros de la secuencia de ruido, diremos que:

+ Ser치 de car치cter aditivo, es decir la se침al que entra al ADC ser치 $s_{R}=s+n$. Siendo $n$ la secuencia que simula la interferencia, y $s$ la senoidal descrita anteriormente.
+ La potencia del ruido ser치 $P_{n}=k_{n}.P_{q}$ siendo el factor $k$ una escala para la potencia del ruido de cuantizaci칩n $P_{q}=\frac{q쑣{12}$.
+ Finalmente, $n$ ser치 incorrelado y Gaussiano.

El ADC que deseamos simular trabajar치 a una frecuencia de muestreo $f_{S}=1000$ Hz y tendr치 un rango anal칩gico de $췀VF=2$ Volts.

Se pide:

    a) Generar el siguiente resultado producto de la experimentaci칩n. B = 4 bits, kn=1.
    b) Analizar para una de las siguientes configuraciones B = 퍆{4, 8 y 16} bits, kn={1/10,1,10}. Discutir los resultados respecto a lo obtenido en a).

Bonus:

    游눑 Encontrar la relaci칩n entre la cantidad de B bits del ADC y el SNR de la se침al digitalizada. (Ver Holton. DSP Principles and App. 6.7 ADC Quantization pag. 387 )
    游뱚 Simular el efecto del aliasing junto con el muestreo.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

# Datos generales
fs = 1000.0
N = 1000

# ADC
B = 4
Vf = 2
q = Vf / 2**B

# Ruido
kn = 10
Pq = q**2 / 12
Pn = kn * Pq

# Ruido gaussiano
n = np.random.normal(0, np.sqrt(Pn), size=N)  # Ruido anal칩gico de la se침al

# Tiempo
ts = 1 / fs
tt = np.arange(0, N) * ts

# Se침al anal칩gica
f0 = fs / N * 10                  # 10 Hz
omega0 = 2 * np.pi * f0
s = np.sqrt(2) * np.sin(omega0 * tt)
sr = s + n                            # Se침al sin cuantizar pero con ruido
srq = q * np.floor(sr / q + 0.5)      # Se침al cuantizada
nq = srq - sr                         # Error de cuantizaci칩n

# FFTs
ff = np.fft.fftfreq(N, d=ts)
ft_Srq = np.fft.fft(srq) / N
ft_SR = np.fft.fft(sr) / N
ft_Nq = np.fft.fft(nq) / N
ft_S = np.fft.fft(s) / N
ft_N = np.fft.fft(n) / N

# C치lculo de potencias promedio
P_s = np.mean(np.abs(ft_S)**2)
P_n = np.mean(np.abs(ft_N)**2)
P_nq = np.mean(np.abs(ft_Nq)**2)

# Acoto a [0, fs/2]
bfrec = (ff >= 0) & (ff <= fs/2)
ff_valid = ff[bfrec]

# Convertir potencias a dB
n_valor_db = 10 * np.log10(2 * P_n)
nq_valor_db = 10 * np.log10(2 * P_nq)
s_valor_db = 10 * np.log10(2 * P_s)

# Se침ales a graficar con estilo de l칤nea
curvas = [
    ('$s_Q$ (Cuantizada)', ff_valid, 10 * np.log10(2 * np.abs(ft_Srq[bfrec])**2), 'b', '-'),
    ('$s_R$ (Sin Cuantizar)', ff_valid, 10 * np.log10(2 * np.abs(ft_SR[bfrec])**2), 'g', '-'),
    ('$n$ ({:.2f} dB)'.format(n_valor_db), ff_valid, [n_valor_db]*len(ff_valid), 'yellow', '-'),     # L칤nea horizontal del ruido anal칩gico
    ('$n_Q$ ({:.2f} dB)'.format(nq_valor_db), ff_valid, [nq_valor_db]*len(ff_valid), 'orange', '-'),  # L칤nea horizontal del error de cuantizaci칩n
    ('$|N(f)|$', ff_valid, 10 * np.log10(2 * np.abs(ft_N[bfrec])**2), 'r', ':'),                 # Curva del ruido anal칩gico
    ('$|n_Q(f)|$', ff_valid, 10 * np.log10(2 * np.abs(ft_Nq[bfrec])**2), 'c', ':'),              # Curva del ruido de cuantizaci칩n
]


# Graficar densidad espectral de potencia
plt.figure(1)
for titulo, frec, pot, color, estilo in curvas:
    plt.plot(frec, pot, label=titulo, color=color, linestyle=estilo)

plt.title('Densidad espectral de potencia')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Potencia [dB]')
plt.grid(True)
plt.xlim([0, fs / 2])
plt.ylim((1.5*np.min(10*np.log10(2*np.array([P_nq, P_n, P_s]))), 10))
plt.legend()

# Gr치fico temporal
plt.figure(2)
plt.plot(tt, srq, lw=1, linestyle='-', label='$s_Q$ (ADC out)')
plt.plot(tt, sr, lw=1, linestyle='-', color='green', label='$s_R$ (ADC in)')
plt.plot(tt, s, lw=1, color='orange', linestyle='-', label='$s$ (analog)')

plt.title(f'Se침al muestreada por un ADC de {B} bits - 췀VF={Vf:.1f} V - q={q:.3f} V')
plt.xlabel('tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.legend()
plt.grid(True)

# Histograma del ruido de cuantizaci칩n
plt.figure(3)
bins = 10
plt.hist(nq, bins=bins, color='lightblue', edgecolor='black')
plt.plot([-q/2, -q/2, q/2, q/2], [0, N/bins, N/bins, 0], '--r')
plt.title(f'Ruido de cuantizaci칩n - {B} bits - 췀VF = {Vf:.1f} V - q = {q:.3f} V')
plt.xlabel('Error de cuantizaci칩n [V]')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.show()
```

### Conclucion:
A partir de los gr치ficos obtenidos, podemos extraer varias observaciones relevantes. Comenzando por el histograma del error de cuantizaci칩n, se aprecia una alta coincidencia con la plantilla te칩rica de distribuci칩n uniforme. Esto confirma que el ruido de cuantizaci칩n presenta efectivamente una distribuci칩n uniforme, como se espera en un ADC ideal.

En cuanto al espectro de potencia, se observa claramente la diferencia entre la se침al cuantizada y la se침al original sin cuantizar. Adem치s, se verifica que el ruido anal칩gico tiene una potencia aproximadamente $k$ veces mayor que la del ruido de cuantizaci칩n, lo cual concuerda con el par치metro configurado ($k = 10$ en este caso).

Por otro lado, al analizar la distribuci칩n espectral del ruido, se nota que la contribuci칩n del ruido total est치 dominada por el ruido anal칩gico. Esto sugiere que, en este caso, el ADC tiene una resoluci칩n superior a la que el sistema realmente necesita, o bien que el sistema anal칩gico introduce un nivel de ruido mayor que el error de cuantizaci칩n. En ambos casos, se concluye que el desempe침o del ADC no es el factor limitante en la calidad final de la se침al digitalizada.

```python
# Par치metros a evaluar
B_list = [4, 8, 16]
kn_list = [0.1, 1, 10]

# Figura para espectros
plt.figure(figsize=(18, 12))
plot_idx = 1

# Colores y estilos fijos para todas las curvas
colores = {
    's_Q': ('b', '-'),
    's_R': ('g', '-'),
    'n': ('yellow', '-'),
    'n_Q': ('orange', '-'),
    '|N(f)|': ('r', ':'),
    '|n_Q(f)|': ('c', ':')
}

for B in B_list:
    q = Vf / 2**B
    Pq = q**2 / 12
    for kn in kn_list:
        Pn = kn * Pq

        # Ruido anal칩gico
        n = np.random.normal(0, np.sqrt(Pn), size=N)
        sr = s + n
        srq = q * np.floor(sr / q + 0.5)
        nq = srq - sr

        # FFTs
        ft_Srq = np.fft.fft(srq) / N
        ft_SR = np.fft.fft(sr) / N
        ft_Nq = np.fft.fft(nq) / N
        ft_N = np.fft.fft(n) / N

        # Potencias promedio
        P_s = np.mean(np.abs(ft_S)**2)
        P_n = np.mean(np.abs(ft_N)**2)
        P_nq = np.mean(np.abs(ft_Nq)**2)

        s_valor_db = 10 * np.log10(2 * P_s)
        n_valor_db = 10 * np.log10(2 * P_n)
        nq_valor_db = 10 * np.log10(2 * P_nq)

        # Subplot para esta configuraci칩n
        plt.subplot(len(B_list), len(kn_list), plot_idx)

        # Curvas con sus etiquetas y valores dB
        curvas = [
            (f'$s_Q$ ({s_valor_db:.2f} dB)', ff_valid, 10 * np.log10(2 * np.abs(ft_Srq[bfrec])**2), 's_Q'),
            (f'$s_R$ ({s_valor_db:.2f} dB)', ff_valid, 10 * np.log10(2 * np.abs(ft_SR[bfrec])**2), 's_R'),
            (f'$n$ ({n_valor_db:.2f} dB)', ff_valid, [n_valor_db]*len(ff_valid), 'n'),
            (f'$n_Q$ ({nq_valor_db:.2f} dB)', ff_valid, [nq_valor_db]*len(ff_valid), 'n_Q'),
            (f'$|N(f)|$', ff_valid, 10 * np.log10(2 * np.abs(ft_N[bfrec])**2), '|N(f)|'),
            (f'$|n_Q(f)|$', ff_valid, 10 * np.log10(2 * np.abs(ft_Nq[bfrec])**2), '|n_Q(f)|'),
        ]

        for titulo, frec, pot, key in curvas:
            color, estilo = colores[key]
            plt.plot(frec, pot, label=titulo, color=color, linestyle=estilo)

        plt.title(f'$B$={B} bits, $k_n$={kn}')
        plt.xlabel('Frecuencia [Hz]')
        plt.ylabel('Potencia [dB]')
        plt.grid(True)
        plt.ylim((1.5*np.min(10*np.log10(2*np.array([P_nq, P_n, P_s]))), 10))
        plt.legend(fontsize='x-small')
        plot_idx += 1

# Ajustar todo
plt.tight_layout()
plt.suptitle('Densidad espectral de potencia para distintas configuraciones de $B$ y $k_n$', fontsize=16, y=1.02)
plt.show()
```

Para este punto se opt칩 por graficar 칰nicamente los espectros de potencia de cada configuraci칩n, ya que el histograma del error de cuantizaci칩n presenta un comportamiento muy similar en todos los casos y el gr치fico temporal no aporta informaci칩n relevante para el an치lisis espectral.

Al observar los resultados, se aprecia una reducci칩n notable del piso de ruido al aumentar la cantidad de bits del ADC. Esto era esperable, ya que tanto el ruido de cuantizaci칩n como el ruido anal칩gico est치n relacionados con la resoluci칩n del convertidor: el primero de forma directa y el segundo de manera forzada en el c칩digo mediante el par치metro $k$.

En la tercera columna (correspondiente a $k=10$), puede verse con claridad el efecto discutido en el punto $a)$ para distintas resoluciones: el ruido anal칩gico domina completamente al de cuantizaci칩n cuando es mucho mayor.

Por el contrario, en la primera columna (con $k=0,1$), se observa el caso inverso: el ruido de cuantizaci칩n es mayor que el ruido anal칩gico, lo que genera un escenario en el que la resoluci칩n del ADC limita significativamente la calidad de la se침al digitalizada.
